```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('raspidat_dummy.csv', parse_dates=['Timestamp'], index_col='Timestamp', date_parser=lambda x: pd.to_datetime(x))
```

```python
df.info()
```

```python
# Find how many duplicate rows exist based on timestamp
dup = df.index.duplicated(keep='first')
print(f"Number of duplicate rows: {dup.sum()}")

# Remove duplicate rows from df and confirm removal
df = df[~df.index.duplicated(keep='first')]
print(f"Number of duplicate rows post removal: {df.index.duplicated(keep='first').sum()}")
```
